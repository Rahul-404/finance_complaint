```python
import pandas as pd
import os
# Read the CSV file
df = pd.read_csv('data/complaints.csv.zip')
```

```python
df.head()
```

```python
file_path = 'data/complaints.parquet'
if not os.path.exists(file_path):
    # Convert to Parquet with Snappy compression
    df.to_parquet(file_path, engine='pyarrow', compression='gzip')
```

```python
# Check the file size of the Parquet file
print(f"Zip file size without Snappy compression: {os.path.getsize('data/complaints.csv.zip') / (1024 ** 2):.2f} MB")
print(f"Parquet file size with Snappy compression: {os.path.getsize('data/complaints.parquet') / (1024 ** 2):.2f} MB")
```
